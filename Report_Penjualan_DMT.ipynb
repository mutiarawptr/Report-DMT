```python
import pandas as pd
```

**Load data exercise**

1. load sheet excel dari file excel
2. Split data dengan memisahkan berdasarkan delimiter ";"
3. Mengatur index dan header

```python
data = pd.read_excel("/content/drive/MyDrive/User Interview for DMT Intern/Pre-Test DMT Intern_October (1).xlsx", sheet_name="Exercise", header=None)
```

```python
data = data[7:].reset_index(drop=True)
data_split = data[0].str.split(';', expand=True)

data_split.head()
```

```python
data_split.columns = data_split.iloc[0]
data = data_split.drop(0).reset_index(drop=True)
data.columns = ['Date / Tanggal', 'Customer Name / Nama Pelanggan', 'Product Code / Kode Barang', 'Description / Nama Barang', 'Barang Terjual']

data.head()
```

**Mengatur format date/tanggal**
1. Convert tipe data kolom Date/Tanggal menjadi datetime

```python
data['Date / Tanggal'] = pd.to_datetime(data['Date / Tanggal'], format='%d-%m-%Y')
data.head()
```

```python
data.info()
```

**Load data Harga Jual**

1. load sheet excel dari file excel
3. Mengatur index dan header

```python
Harga_jual = pd.read_excel("/content/drive/MyDrive/User Interview for DMT Intern/Pre-Test DMT Intern_October (1).xlsx", sheet_name="Harga Jual Product", header=None)

Harga_jual.columns = Harga_jual.iloc[0]
Harga_jual = Harga_jual.drop(0).reset_index(drop=True)
Harga_jual.head()
```

```python
Harga_jual.info()
```

**Merge data dan harga jual**

1. Merge dataframe *data* dan *harga_jual* mengacu pada kolom 'Product Code / Kode Barang' yang terdapat pada dataframe *data*

```python
Data_merge = pd.merge(data, Harga_jual, on='Product Code / Kode Barang', how='left')
Data_merge
```

**Menghitung total harga jual**

1. Ubah kedua jenis kolom menjadi integer
2. Buat kolom baru sebagai hasil perkalian antara kolom Barang Terjual dan Harga Jual

```python
Data_merge['Barang Terjual'] = Data_merge['Barang Terjual'].astype(int)
Data_merge['Harga Jual'] = Data_merge['Harga Jual'].astype(int)

Data_merge['total harga jual'] = Data_merge['Barang Terjual'] * Data_merge['Harga Jual']
Data_merge
```
